```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(train, y, test_size=0.33, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier(max_depth=5)
dtc.fit(X_train, y_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
```

```python
dtc = DecisionTreeClassifier(max_depth=5)
dtc.fit(X_train, y_train)

knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)

lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
lr.coef_
```

```python
df=pd.DataFrame(list(zip(lr.coef_[0],train.columns)))
df.sort_values(0)
```

```python
import math
```

```python
def predict(features,coefs,lasr_coef):
    s=0
    for x1,x2 in zip(features, coefs):
        s+=x1*x2
    s+=lasr_coef
    return (1/(1+math.e**(-1*s)))
```

```python
predict([1,1],[1,1],1)
```

```python
lr.intercept_ #свободный член
```

```python
predict(list(X_test.iloc[0]),lr.coef_[0],lr.intercept_[0])
```
